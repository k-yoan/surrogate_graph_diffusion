```python
import torch_geometric as pyg
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from random import randint

%matplotlib inline
```

## Facebook Dataset

```python
facebook = pd.read_csv(
    "data/facebook_combined.txt.gz",
    compression="gzip",
    sep=" ",
    names=["start_node", "end_node"],
)

''' 
Downloaded from http://snap.stanford.edu/data/ego-Facebook.html

'''
```

```python
G = nx.from_pandas_edgelist(facebook, "start_node", "end_node")
G.number_of_nodes() #4039
```

```python

pos = nx.spring_layout(G, iterations=15, seed=1721)
fig, ax = plt.subplots(figsize=(15, 9))
ax.axis("off")
plot_options = {"node_size": 10, "with_labels": False, "width": 0.15}
nx.draw_networkx(G, pos=pos, ax=ax, **plot_options)
```

```python
colors = ["" for x in range(G.number_of_nodes())]
for com in nx.community.asyn_fluidc(G, 4, seed=0):
    color = "#%06X" % randint(0, 0xFFFFFF)  # creates random RGB color
    for node in list(com):
        colors[node] = color
```

```python
plt.figure(figsize=(15, 9))
plt.axis("off")
nx.draw_networkx(
    G, pos=pos, node_size=10, with_labels=False, width=0.15, node_color=colors
)
```

## Twitter Dataset

```python
twitter = pd.read_csv(
    "data/twitter_combined.txt.gz",
    compression="gzip",
    sep=" ",
    names=["start_node", "end_node"],
)

''' 
Downloaded from http://snap.stanford.edu/data/ego-Twitter.html

'''
G = nx.from_pandas_edgelist(twitter, "start_node", "end_node")
G.number_of_nodes() #81306
```

```python

pos = nx.spring_layout(G, iterations=2, seed=1721)
fig, ax = plt.subplots(figsize=(15, 9))
ax.axis("off")
plot_options = {"node_size": 10, "with_labels": False, "width": 0.15}
nx.draw_networkx(G, pos=pos, ax=ax, **plot_options)
## The number of nodes is way greater than Facebook's
```
